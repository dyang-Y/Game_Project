```python
from selenium import webdriver
```

```python
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()

driver.get('https://store.steampowered.com/charts/topselling/KR')
```

```python
driver.find_elements(By.CSS_SELECTOR, 'table._3arZn0BMPzyhcYNADe193m')
```

```python
element = driver.find_element(By.CSS_SELECTOR, 'tr._2-RN6nWOY56sNmcDHu069P')
print(element.text)
```

```python
elements = driver.find_elements(By.CSS_SELECTOR, 'tr._2-RN6nWOY56sNmcDHu069P')

for element in elements:
    rank = element.find_element(By.CSS_SELECTOR, 'td._34h48M_x9S-9Q2FFPX_CcU').text
    title = element.find_element(By.CSS_SELECTOR, 'div._1n_4-zvf0n4aqGEksbgW9N').text
    price = element.find_element(By.CSS_SELECTOR, 'div._3j4dI1yA7cRfCvK8h406OB').text
    print(rank, title, f"| {price}")
```
